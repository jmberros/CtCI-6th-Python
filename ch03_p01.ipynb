# Solution 1: fixed size array

```python
class FullStackException(Exception):
    pass

class EmptyStackException(Exception):
    pass

class FixedMultiStack:
    # TODO: add class docstring

    def __repr__(self) -> str:
        values_str = [str(val) if val is not None else "_" for val in self.values]
        return f" ".join(values_str)

    def __init__(self, num_stacks: int, stack_capacity: int):
        self.n_stacks = num_stacks
        self.stack_capacity = stack_capacity

        self.values = [None] * self.n_stacks * stack_capacity
        self.stack_sizes = [0] * self.n_stacks

    def push(self, item, stack_num: int):
        offset = stack_num * self.stack_capacity
        stack_size = self.stack_sizes[stack_num]

        if self.is_full(stack_num):
            raise FullStackException

        top_index = offset + stack_size
        self.values[top_index] = item
        self.stack_sizes[stack_num] += 1

    def pop(self, stack_num: int):
        if self.is_empty(stack_num):
            raise EmptyStackException

        i = self.top_index(stack_num)
        value = self.values[i]
        self.values[i] = None
        self.stack_sizes[stack_num] -= 1

        return value

    def peek(self, stack_num: int):
        if self.is_empty(stack_num):
            raise EmptyStackException
        
        return self.values[self.top_index(stack_num)]

    def is_full(self, stack_num: int) -> bool:
        return self.stack_sizes[stack_num] == self.stack_capacity

    def is_empty(self, stack_num: int) -> bool:
        return self.stack_sizes[stack_num] == 0

    def top_index(self, stack_num: int) -> int:
        """Returns the index of the top of the stack."""  # TODO: Args + Returns
        offset = stack_num * self.stack_capacity
        return offset + self.stack_sizes[stack_num] - 1

###

stack = FixedMultiStack(num_stacks=3, stack_capacity=5)

for i in range(5):
    stack.push(i + 1, stack_num=0)
    print(stack)

for i in range(3):
    stack.pop(stack_num=0)
    print(stack)

print(stack)

stack.peek(0)
```

# Flexible size of stacks

```python
class FlexibleMultiStack:
    # TODO: add class docstring
    # TODO: docstrings to the methods

    def __init__(self, num_stacks: int, stack_capacity: int):
        self.n_stacks = num_stacks
        self.stack_capacity = stack_capacity
        self.values = [None] * self.n_stacks * stack_capacity
        self.stack_sizes = [0] * self.n_stacks

    def __repr__(self) -> str:
        values_str = [str(val) if val is not None else "_" for val in self.values]
        for i in range(self.n_stacks - 1):
            j = self.offset(i) + self.stack_capacity - 1
            values_str[j] += " |"
        return f" ".join(values_str)

    def increase_capacity(self):
        new_capacity = self.stack_capacity * 2
        new_values = [None] * self.n_stacks * new_capacity
        for stack_num in range(self.n_stacks):
            i = self.offset(stack_num)
            j = i + self.stack_capacity
            new_offset = new_capacity * stack_num
            new_values[new_offset:new_offset + self.stack_capacity] = self.values[i:j]
        self.values = new_values
        self.stack_capacity = new_capacity

    def push(self, item, stack_num: int):
        stack_size = self.stack_sizes[stack_num]
        if self.is_full(stack_num):
            self.increase_capacity()
        self.values[self.top_index(stack_num) + 1] = item
        self.stack_sizes[stack_num] += 1

    def pop(self, stack_num: int):
        if self.is_empty(stack_num):
            raise EmptyStackException
        i = self.top_index(stack_num)
        value = self.values[i]
        self.values[i] = None
        self.stack_sizes[stack_num] -= 1

        return value

    def peek(self, stack_num: int):
        if self.is_empty(stack_num):
            raise EmptyStackException
        return self.values[self.top_index(stack_num)]

    def is_full(self, stack_num: int) -> bool:
        return self.stack_sizes[stack_num] == self.stack_capacity

    def is_empty(self, stack_num: int) -> bool:
        return self.stack_sizes[stack_num] == 0
    
    def offset(self, stack_num: int) -> int:
        return stack_num * self.stack_capacity

    def top_index(self, stack_num: int) -> int:
        """Returns the index of the top of the stack."""
        return self.offset(stack_num) + self.stack_sizes[stack_num] - 1


stack = FlexibleMultiStack(num_stacks=3, stack_capacity=2)

stack.push(1, stack_num=1)
stack.push(2, stack_num=1)

stack.push("A", stack_num=0)
stack.push("B", stack_num=0)

stack.push("Y", stack_num=2)
stack.push("Z", stack_num=2)

print(stack)

stack.push(3, stack_num=1)
stack.push("X", 2)
stack.push("R", 2)
stack.push("T", 2)
stack.push("S", 2)

print(stack)
```
